```python
import subprocess

def run_git(args):
    try:
        result = subprocess.run(
            ["git"] + args,
            capture_output=True,
            text=True,
            check=False
        )
        print(f"CMD: git {' '.join(args)}")
        print("STDOUT:", result.stdout)
        print("STDERR:", result.stderr)
        return result.returncode
    except Exception as e:
        print(f"Failed to run git: {e}")
        return -1

run_git(["status"])
```

```python
# Stage and commit
run_git(["add", "run_tests.py"])
run_git(["commit", "-m", "Update test runner script with diagnostic tools"])
run_git(["push"])
```

```python
# Check ChromaDB Index Status
import sys
from pathlib import Path
# Ensure src is in path
sys.path.append(str(Path.cwd() / "src"))

from ordinis.rag.vectordb.chroma_client import ChromaClient

try:
    client = ChromaClient()
    
    print("--- Knowledge Base Status ---")
    
    # Check Text Collection
    try:
        text_coll = client.get_text_collection()
        count = text_coll.count()
        print(f"Text Documents: {count}")
        if count > 0:
            print(f"Sample Text: {text_coll.peek(1)['documents'][0][:100]}...")
    except Exception as e:
        print(f"Text Collection: Not found or error ({e})")

    # Check Code Collection
    try:
        code_coll = client.get_code_collection()
        count = code_coll.count()
        print(f"Code Documents: {count}")
        if count > 0:
            print(f"Sample Code: {code_coll.peek(1)['documents'][0][:100]}...")
    except Exception as e:
        print(f"Code Collection: Not found or error ({e})")
        
except Exception as e:
    print(f"Failed to connect to ChromaDB: {e}")
```
